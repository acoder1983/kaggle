```python


slices=listdir(i_path)
tracks=set([''])
for i in tnrange(len(slices)):
    s=slices[i]
    with open(join(i_path,s)) as f:
        plist=json.load(f)
        for p in plist['playlists']:
            for t in p['tracks']:
                tracks.add(t['track_uri'])
tracks.remove('')
len(tracks)
```

```python
tracks=list(tracks)
s='team_info,main,acodersop,acodersop@gmail.com\n'
with open('raw/challenge/challenge_set.json') as f:
    cha_set=json.load(f)
    playlists=cha_set['playlists']
    for i in tnrange(len(playlists)):
        p=playlists[i]
        s+=str(p['pid'])+','
        idxs=np.random.permutation(len(tracks))[:1000]
        k=0
        seed_tracks=[t['track_uri'] for t in p['tracks']]
        for j in idxs:
            if tracks[j] not in seed_tracks:
                s+=tracks[j]+','
                k+=1
            if k==500:
                break
        s=s[:-1]+'\n'
with open('output/random.csv','w') as f:
    f.write(s)
```

```python
import re
pattern = re.compile('[\W_]+')
# names=[]
def build_title(name):
    n=str(name).lower()
    words=n.split()
    for i,w in enumerate(words):
        words[i]=pattern.sub('', w)
    w=''.join(words)
#     if len(w)==0:
#         names.append(name)
    return w
playlists['title']=playlists.name.apply(build_title)
playlists_tst['title']=playlists_tst.name.apply(build_title)
title_tra=set(playlists.title)
title_tst=set(playlists_tst.title)
not_in_titles=title_tst-title_tra&title_tst
len(not_in_titles),not_in_titles
```
